## MLPRegressor

Különböző beállításokkal,

Baromira nem mindegy, hogy milyen beállításokat használunk, néha ezek hatására képtelen tanulni.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.arange(0, 600, 1)

y = np.sin(x/180) + np.sin(x/60)

x = x.reshape(-1,1)
y = y
```

## Tanh

```python
mlp = MLPRegressor(hidden_layer_sizes=(3, ),
                   activation='tanh', 
                   solver='sgd',
                   batch_size='auto',
                   learning_rate_init=0.01,
                   max_iter=2000,
                   shuffle=False,
                   random_state=1,
                   momentum=0.9,
                   nesterovs_momentum=True,
                   early_stopping=True,
                   n_iter_no_change=2000)
```

## Ilyen a story ha az x nincs normalizálva

A bement tarománya 0-600

```python
mlp.fit(x, y)
predicted = mlp.predict(x)
plt.scatter(x, y)
plt.plot(x, predicted, c='black')
plt.show()
```

## Ilyen az amikor x és y 0,1 közé van normalizálva

```python
scaler_x_one = MinMaxScaler(feature_range=(0,1))
scaler_x_one.fit(x)
xs1 = scaler_x_one.transform(x)
print('xs1.min = ', xs1.min())
print('xs1.max = ', xs1.max())
scaler_y_one = MinMaxScaler(feature_range=(0,1))
scaler_y_one.fit(y.reshape(-1,1))
ys1 = scaler_y_one.transform(y.reshape(-1,1))
print('ys1.min = ', ys1.min())
print('ys1.max = ', ys1.max())
print('ys1.shape = ', ys1.shape)
x_scaled = xs1
y_scaled = ys1
mlp.fit(x_scaled, y_scaled)
predicted = mlp.predict(x_scaled)
plt.scatter(x_scaled, y_scaled)
plt.plot(x_scaled, predicted, c='black')
plt.show()
```

## Ilyen az amikor x (-1,1) közé y (0,1) közé van normalizálva

```python
scaler_x_one = MinMaxScaler(feature_range=(-1,1))
scaler_x_one.fit(x)
xs1 = scaler_x_one.transform(x)
print('xs1.min = ', xs1.min())
print('xs1.max = ', xs1.max())
scaler_y_one = MinMaxScaler(feature_range=(0,1))
scaler_y_one.fit(y.reshape(-1,1))
ys1 = scaler_y_one.transform(y.reshape(-1,1))
print('ys1.min = ', ys1.min())
print('ys1.max = ', ys1.max())
print('ys1.shape = ', ys1.shape)
x_scaled = xs1
y_scaled = ys1
mlp.fit(x_scaled, y_scaled)
predicted = mlp.predict(x_scaled)
plt.scatter(x_scaled, y_scaled)
plt.plot(x_scaled, predicted, c='black')
plt.show()
```

## Logistic átviteli függvény Adam optimizerrel x és y (-1,1) közé transzformálva

```python
mlp = MLPRegressor(hidden_layer_sizes=(3, ),
                   activation='logistic', 
                   solver='adam',
                   batch_size='auto',
                   learning_rate_init=0.01,
                   max_iter=2000,
                   shuffle=True,
                   random_state=1,
                   momentum=0.9,
                   nesterovs_momentum=True,
                   early_stopping=True,
                   n_iter_no_change=2000)

scaler_x_one = MinMaxScaler(feature_range=(-1,1))
scaler_x_one.fit(x)
xs1 = scaler_x_one.transform(x)
print('xs1.min = ', xs1.min())
print('xs1.max = ', xs1.max())
scaler_y_one = MinMaxScaler(feature_range=(-1,1))
scaler_y_one.fit(y.reshape(-1,1))
ys1 = scaler_y_one.transform(y.reshape(-1,1))
print('ys1.min = ', ys1.min())
print('ys1.max = ', ys1.max())
print('ys1.shape = ', ys1.shape)
x_scaled = xs1
y_scaled = ys1
mlp.fit(x_scaled, y_scaled)
predicted = mlp.predict(x_scaled)
plt.scatter(x_scaled, y_scaled)
plt.plot(x_scaled, predicted, c='black')
plt.show()
```
