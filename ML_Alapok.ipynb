## Linear Regression

Placheholder

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.arange(0, 100, 5)

y = x * 1.2 + 3
```

```python
regression = LinearRegression()

regression.fit(x.reshape(-1,1),y)

predicted = regression.predict(x.reshape(-1, 1))

estimated = regression.intercept_ + regression.coef_ * x
```

```python
plt.figure(figsize=(12, 5))
plt.scatter(x, y, label = 'y')
plt.plot(x, predicted, label = 'predicted')
plt.plot(x, estimated, label = 'estimated y = ' + str(round(regression.intercept_, 3)) + ' + ' + str(regression.coef_) + ' * x')
plt.legend(frameon = False)
plt.show()
```
